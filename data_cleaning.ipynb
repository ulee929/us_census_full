```python
import pandas as pd
pd.options.display.max_columns=50
```

```python
# Read data
df_income_learn = pd.read_csv('census_income_learn.csv', header=None)
df_income_test = pd.read_csv('census_income_test.csv', header=None)
```

```python
print(f"Train set size: {len(df_income_learn)}")
print(f"Test set size: {len(df_income_test)}")
```

```python
# Set column names
df_columns = ['age',
'class_of_worker',
'detailed_industry_recode',
'detailed_occupation_recode',
'education',
'wage_per_hour',
'enroll_in_edu_inst_last_wk',
'marital_stat',
'major_industry_code',
'major_occupation_code',
'race',
'hispanic_origin',
'sex',
'member_of_a_labor_union',
'reason_for_unemployment',
'full_or_part_time_employment_stat',
'capital_gains',
'capital_losses',
'dividends_from_stocks',
'tax_filer_stat',
'region_of_previous_residence',
'state_of_previous_residence',
'detailed_household_and_family_stat',
'detailed_household_summary_in_household',
'instance_weight',
'migration_code-change_in_msa',
'migration_code-change_in_reg',
'migration_code-move_within_reg',
'live_in_this_house_1_year_ago',
'migration_prev_res_in_sunbelt',
'num_persons_worked_for_employer',
'family_members_under_18',
'country_of_birth_father',
'country_of_birth_mother',
'country_of_birth_self',
'citizenship',
'own_business_or_self_employed',
'fill_inc_questionnaire_for_veterans_admin',
'veterans_benefits',
'weeks_worked_in_year',
'year',
'income']
```

```python
df_income_learn.columns = df_columns
df_income_test.columns = df_columns
```

```python
# Drop instance weight column
df_income_learn.drop(['instance_weight'], axis=1, inplace=True)
df_income_test.drop(['instance_weight'], axis=1, inplace=True)
print(len(df_income_learn.columns))
```

```python
# Find duplicate rows for train and test
duplicated_learn = df_income_learn[df_income_learn.duplicated()]
print(len(duplicated_learn))
```

```python
duplicated_test = df_income_test[df_income_test.duplicated()]
print(len(duplicated_test))
```

```python
# Drop duplicate rows
df_income_learn.drop_duplicates(keep=False,inplace=True)
print(f"Train set size after dropping duplicates: {len(df_income_learn)}")
```

```python
df_income_test.drop_duplicates(keep=False,inplace=True)
print(f"Test set size after dropping duplicates: {len(df_income_test)}")
```

```python
# Check for null values
df_income_learn.isnull().sum()
```

```python
df_income_test.isnull().sum()
```

```python
# Look at data after dropping duplicates
df_income_learn.describe()
```

```python
df_income_test.describe()
```

```python
# Replace income values with binary integer values
df_income_learn['income'] = df_income_learn['income'].replace(" - 50000.", 0) 
df_income_learn['income'] = df_income_learn['income'].replace(" 50000+.", 1)
df_income_learn.income.value_counts()
```

```python
df_income_test['income'] = df_income_test['income'].replace(" - 50000.", 0) 
df_income_test['income'] = df_income_test['income'].replace(" 50000+.", 1)
df_income_test.income.value_counts()
```
