#### Drop unnecessary data
We do not need the instance_weight column, duplicated rows, or the detailed_industry_recode and detailed_occupation_recode columns

```python
# Drop instance weight column
df_income_learn.drop(['instance_weight'], axis=1, inplace=True)
df_income_test.drop(['instance_weight'], axis=1, inplace=True)
print(len(df_income_learn.columns))
```

```python
# Find duplicate rows for train and test
duplicated_learn = df_income_learn[df_income_learn.duplicated()]
print(len(duplicated_learn))
```

```python
duplicated_test = df_income_test[df_income_test.duplicated()]
print(len(duplicated_test))
```

```python
# Drop duplicate rows
df_income_learn.drop_duplicates(keep=False,inplace=True)
print(f"Train set size after dropping duplicates: {len(df_income_learn)}")
```

```python
df_income_test.drop_duplicates(keep=False,inplace=True)
print(f"Test set size after dropping duplicates: {len(df_income_test)}")
```

```python
# Look at data after dropping duplicates
df_income_learn.describe()
```

```python
df_income_test.describe()
```

**Dropping duplicates seems to be a critical step in data cleaning. For example, after dropping duplicates, the mean age increased from 34.49 to 40.05. This suggests that leaving duplicates in for analysis would have skewed results toward the younger population.**

```python
# Drop detailed_industry_recode and detailed_occupation_recode columns since they are duplicates of major_industry_code and major_occupation_code
df_income_learn.drop(['detailed_industry_recode'], axis=1, inplace=True)
df_income_learn.drop(['detailed_occupation_recode'], axis=1, inplace=True)

df_income_test.drop(['detailed_industry_recode'], axis=1, inplace=True)
df_income_test.drop(['detailed_occupation_recode'], axis=1, inplace=True)
```

### We can bin the education column to reduce the number of categories

```python
df_income_learn['education'].value_counts()
```

**We can bin education to up to 6th grade and high school without graduating**

```python
up_to_6th_grade = [' 5th or 6th grade', ' 1st 2nd 3rd or 4th grade', ' Less than 1st grade']
high_school_no_graduate = [' 9th grade', ' 10th grade', ' 11th grade', ' 12th grade no diploma']

for column in up_to_6th_grade:
    df_income_learn['education'] = df_income_learn['education'].replace(column, 'up_to_6th_grade')

for column in high_school_no_graduate:
    df_income_learn['education'] = df_income_learn['education'].replace(column, 'high_school_no_graduate')
```

```python
df_income_learn['education'].value_counts()
```

```python
for column in up_to_6th_grade:
    df_income_test['education'] = df_income_test['education'].replace(column, 'up_to_6th_grade')
for column in high_school_no_graduate:
    df_income_test['education'] = df_income_test['education'].replace(column, 'high_school_no_graduate')
```

```python
df_income_test['education'].value_counts()
```

### Finally, replace the income label with binary integer values

```python
# Replace income values with binary integer values
df_income_learn['income'] = df_income_learn['income'].replace(" - 50000.", 0) 
df_income_learn['income'] = df_income_learn['income'].replace(" 50000+.", 1)
df_income_learn.income.value_counts()
```

```python
df_income_test['income'] = df_income_test['income'].replace(" - 50000.", 0) 
df_income_test['income'] = df_income_test['income'].replace(" 50000+.", 1)
df_income_test.income.value_counts()
```
